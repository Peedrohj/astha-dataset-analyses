## Um grÃ¡fico de densidade dos dados

Na parte de densidade resolvemos criar um grÃ¡fico para uma das colunas, atravÃ©s dele Ã© possÃ­vel analisar a variaÃ§Ã£o nos valores escolhidos.


```python
def density(dataframe):
    for column in dataframe.columns:
            dataframe[[column]].plot(kind='density') 
    
```

```python
density(reduced_df)
```
