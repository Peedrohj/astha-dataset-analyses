## Redução do dataframe para usar apenas as colunas selecionadas

Por ser um dataset muito grande, contendo 599 variáveis, decimos criar um filtro e selecionar os valores mais importantes que indentificamos na base, são eles:

1-(S8Q7R) Nos últimos 3 meses, tomou medicamento para asma com receita com um inalador?

2-(S7Q14) Tem carpete ou tapete no seu quarto?

3-(S7Q11) Algum profissional de saúde lhe recomendou alterar coisas na casa, na escola ou no trabalho  para melhorar a asma?

4-(S6Q1) Um médico ou outro profissional de saúde já ensinou você , seus pais ou tutor a reconhecer sinais ou sintomas de um episódio de asma ?

5-(S6Q2) Um médico ou outro profissional de saúde já ensinou você,  seus pais ou tutor o que fazer durante um episódio ou ataque de asma?


6-(S7Q18P)PROXY. Na sua vida inteira ja fumou no mínimo 100 cigarros?

7-(S7Q19P)PROXY. Fuma cigarros todo dia, alguns dias ou nunca?


8-(S7Q6) Seu domicílio tem animais de estimação, como cães, gatos, hamsters, pássaros ou outros animais de estimação com penas ou peludos, que são mantidos dentro de casa?


9-(S5Q1R) Nos últimos 12 meses, quantas vezes  consultou um médico ou outro profissional de saúde para uma verificação de rotina da asma?

10-(S5Q11R) Nos últimos 12 meses, quantos dias de daycare / pré-escola / escola  ou trabalho por causa de asma perdeu?

11-(S7Q17 ) O animal de estimação  é permitido no quarto?

12-(S8Q3 ) Um profissional de saúde mostrou você, seus pais ou tutores como usar o inalador?

13-(S7Q6P)PROXY. Seu domicílio tem animais de estimação, como cães, gatos, hamsters, pássaros ou outros animais de estimação com penas ou peludos, que são mantidos dentro de casa?

```python
chosed_columns = ["S8Q7R", "S7Q14", "S7Q11", "S6Q1", "S6Q2", "S7Q18P", "S7Q19P", "S7Q6", "S5Q1R", "S5Q11R", "S7Q17", "S8Q3", "S7Q6P"]

reduced_df = df[chosed_columns]
```

## Um gráfico de densidade dos dados

Na parte de densidade resolvemos criar um gráfico para uma das colunas, através dele é possível analisar a variação nos valores escolhidos.


```python
def density(dataframe):
    for column in dataframe.columns:
            dataframe[[column]].plot(kind='density') 
    
```

```python
density(reduced_df)
```

## Uma análise de uma amostra dos dados

Para realizar a análise das amostras a base foi dividida em 4 grupos, sendo cada um deles o dobro do valor do seu antecessor, ou seja, o primeiro grupo com 500 amostras, o segundo com 1000, o terceiro com 2000 e o ultimo com 4000 mil amostras. A partir dessa divsão foi analisado a quantidade de valores nulos existentes em cada uma das colunas selecionadas anteriormente no reduced_df depois foram plotados gráficos de dispersão para cada uma das colunas.

Algumas diferenças ficaram visíveis entre os gráficos, principalmente quando comparando a menor amostra(500 itens), com a maior amostra(4000 itens), a mais gritante, foi a diferença entre a dispersão do eixo X, que é representação da distância entre os valores escolhidos. Essa diferença fica clara logo quando observamos o primeiro gráfico da variável (S8Q7R), na primeira amostra o menor valor no eixo x é 0.5 e o maior é 1.5, enquanto no ultimo gráfico o maior valor chega a ser oito 

variável (S8Q7R) -> Nos últimos 3 meses, tomou medicamento para asma com receita com um inalador?



```python
sample_df_500 = reduced_df.sample(n=500)
sample_df_1000 = reduced_df.sample(n=1000)
sample_df_2000 = reduced_df.sample(n=2000)
sample_df_4000 = reduced_df.sample(n=4000)
```
